```python
# Check Missing Values
import pandas as pd

df = pd.read_csv("data.csv", encoding="latin1")
df.isnull().sum()
```

```python
# structure of the data
df.info()
```

```python
# Count of missing values in each column
df.isnull().sum()
```

```python
# percentage of missing values
df.isnull().mean() * 100
```

```python
# Check for NEGATIVE QUANTITY
df[df['Quantity'] < 0].head()
df[df['Quantity'] < 0].shape[0]

```

```python
#Check for ZERO or NEGATIVE PRICE
df[df['UnitPrice'] <= 0].head()
df[df['UnitPrice'] <= 0].shape[0]
```

```python
# Check unusual StockCodes
df[~df['StockCode'].str.isnumeric()].head()
```

```python
# Check if InvoiceDate is valid
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

df['InvoiceDate'].isnull().sum()
```

```python
# Check for DUPLICATE ROWS
df.duplicated().sum()
df[df.duplicated()].head()
```

```python
# Check if same StockCode has multiple Descriptions
df.groupby('StockCode')['Description'].nunique().sort_values(ascending=False).head(10)
```

```python
# Find canceled invoices
df[df['InvoiceNo'].astype(str).str.startswith('C')].shape[0]
```

```python
# Check OUTLIERS
df[df['Quantity'] > 1000].shape[0]
df[df['UnitPrice'] > 500].shape[0]
```
